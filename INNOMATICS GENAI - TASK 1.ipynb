```python
# Data Science Internship – Logic Building Task 1
# Author: HIMA SHAJU
# Date: 12/2/2026

# 1. User Login Check
print("Task 1: User Login Check")
username1 = input("Enter username: ")
password1 = input("Enter password: ")
username = "hima"
password = "amih"

if username1 == username and password1 == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
print("\n")


# 2. Pass or Fail Analyzer
print("Pass or Fail Analyzer")
marks = [33, 65, 93, 45, 70]
passed = 0
fail = 0
for mark in marks:
    if mark>= 50:
        passed+= 1
    else:
        fail+= 1
print("Total Students Passed:", passed)
print("Total Students Failed:", fail)
print("\n")


# 3. Data Cleaner
print("Data Cleaner")
names = [" Hima ", "Abin", " ANGEL "]
cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)
print("Cleaned Names:", cleaned_names)
print("\n")


# 4. Message Length Analyzer
print("Message Length Analyzer")
messages = ["Hello", "Welcoming you to the platform", "OK"]
for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    if length > 10:
        print("⚠ This message is longer than 10 characters")
print("\n")


# 5. Error Message Detector
print("Error Message Detector")
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error = logs.count("ERROR")
print("Total ERROR entries:", error)
```
