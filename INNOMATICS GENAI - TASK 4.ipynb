```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge amount must be at least ₹50.")
        elif amount not in valid_plans:
            print("Invalid plan. Available plans:", valid_plans)
        else:
            print("Recharge successful for ₹", amount)
            break

validate_recharge()
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Sample Data
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System
def process_result(marks_list):
    total = 0
    for mark in marks_list:
        total += mark

    average = total / len(marks_list)

    if average >= 50:
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"

# Example
marks = [60, 70, 45, 55]
print(process_result(marks))
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare():
    base_fare = 50
    rate_per_km = 12

    while True:
        km = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        fare = base_fare + (km * rate_per_km)

        if peak == "yes":
            fare += fare * 0.25

        print(f"Total Fare: ₹{fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break

calculate_fare()
```

```python
#Problem 5: Employee Attendance Eligibility Checker
def check_eligibility(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% - Eligible"
    else:
        return f"Attendance: {percentage:.2f}% - Not Eligible"

# Example
attendance = ["P", "A", "P", "P", "A", "P", "P"]
print(check_eligibility(attendance))
```

```python
#Problem 6: Password Strength Checker
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password: Minimum length should be 8"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password: Must contain at least one digit and one special character (@#$)"

# Example
pwd = input("Enter password: ")
print(check_password(pwd))
```
