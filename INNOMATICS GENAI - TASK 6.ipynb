```python
#Problem 1: Smart Parking Lot Management System
def smart_parking_system(capacity, vehicle_logs):
    
    parked = 0
    peak_usage = 0
    
    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        # Track peak usage
        if parked > peak_usage:
            peak_usage = parked
    
    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak_usage)
    
    if parked > capacity:
        print("Parking Status: Over Capacity Alert!")
    else:
        print("Parking Status: Available")
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
#Problem 2: Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic_level, weather):
    # Base time calculation (5 minutes per km)
    eta = distance * 5
    
    # Traffic adjustments
    if traffic_level.lower() == "high":
        eta += 15
    elif traffic_level.lower() == "medium":
        eta += 8
    
    # Weather adjustments
    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "stormy":
        eta += 20
    
    print("Estimated Delivery Time:", eta, "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

```python
#Problem 3: Movie Theatre Seat Occupancy Analyzer
def theatre_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100
    
    print("Occupancy:", round(occupancy_percentage), "%")
    
    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
    else:
        print("Show Status: Seats Available")
booked = [1] * 150
theatre_occupancy(200, booked)
```

```python
#Problem 4: Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total = 0
    
    for reading in cpu_readings:
        total += reading
    
    average = total / len(cpu_readings)
    
    print("Average CPU Load:", round(average), "%")
    
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
classify_server_load([45, 60, 70, 85, 90])
```

```python
#Problem 5: Smart Classroom Resource Usage Monitor
def classroom_resource_monitor(resource_usage):
    overused = []
    
    # Threshold = more than 8 hours usage
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resources)
```

```python
#Problem 6: Online Event Registration Capacity Controller
def event_registration_controller(capacity, registrations):
    
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
event_registration_controller(100, 105)
```
